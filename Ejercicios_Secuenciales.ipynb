```python
# 02 Promedio de cuatro exámenes
# Definición de variables
e1 = float(input("Ingrese nota 1: "))
e2 = float(input("Ingrese nota 2: "))
e3 = float(input("Ingrese nota 3: "))
e4 = float(input("Ingrese nota 4: "))
# Proceso de datos
promedio = (e1 + e2 + e3 + e4) / 4
# Salida de resultados
print("Promedio:", promedio)

```

```python
# 03. Porcentaje de inversión de tres personas
# Definción de variables
p1 = float(input("Inversión persona 1: "))
p2 = float(input("Inversión persona 2: "))
p3 = float(input("Inversión persona 3: "))
# Cálculo
total = p1 + p2 + p3
porc1 = p1 * 100 / total
porc2 = p2 * 100 / total
porc3 = p3 * 100 / total
# Salida de Resultados
print("Persona 1:", porc1, "%")
print("Persona 2:", porc2, "%")
print("Persona 3:", porc3, "%")
```

```python
# 04. Sueldo con comisiones
# Definición de variables
sueldo_base = float(input("Ingrese sueldo base: "))
# Definición de variables e ingreso de datos
venta1 = float(input("Venta 1: "))
venta2 = float(input("Venta 2: "))
venta3 = float(input("Venta 3: "))
# Cálculo de datos
total_ventas = venta1 + venta2 + venta3
comision = total_ventas * 0.25
sueldo_total = sueldo_base + comision
# Salida de resultados
print("Comisión:", comision)
print("Sueldo total:", sueldo_total)
```

```python
# 05. Inversión bancaria al 2% mensual
# Definición de Variables e ingreso de datos
capital = float(input("Ingrese capital: "))
# Cálculo de datos
monto_final = capital * (1 + 0.02) ** 12
ganancia = monto_final – capital
# Salida de resultados
print("Monto final:", round(monto_final, 2))
print("Ganancia:", round(ganancia, 2))
```

```python
# 06. Tiempo vivido por una persona
# Definición de Variables e ingreso de datos
edad = int(input("Ingrese su edad en años: "))
# Cálculo de datos
meses = edad * 12
semanas = edad * 52
dias = edad * 365
horas = dias * 24
# Salida de resultados
print("Meses vividos:", meses)
print("Semanas vividas:", semanas)
print("Días vividos:", dias)
print("Horas vividas:", horas)
```

```python
# 07. Cobro de estacionamiento
# Definición de Variables e ingreso de datos
horas = int(input("Horas utilizadas: "))
minutos = int(input("Minutos adicionales: "))
# Cálculo de datos
costo_horas = horas * 5.00
costo_minutos = minutos * 0.10
total = costo_horas + costo_minutos
# Salida de resultados
print("Monto a pagar: S/", round(total, 2))
```

```python
# 08 Programa para calcular el área de un terreno (La forma A)
# Entrada de datos
A = float(input("Ingrese el lado A: "))
B = float(input("Ingrese el lado B: "))
C = float(input("Ingrese el lado C: "))
# Cálculo de áreas parciales
area_rectangulo = B * C
area_triangulo = (B * (A - C)) / 2

area_total = area_rectangulo + area_triangulo
# Resultados
print("\nRESULTADOS")
print("Área del rectángulo:", area_rectangulo)
print("Área del triángulo:", area_triangulo)
print("Área total del terreno:", area_total)
```

```python
# 09. Compra con descuento y cambio
# Entrada de datos
monto = float(input("Monto de compra: "))
descuento = float(input("Porcentaje de descuento: "))
pago = float(input("Dinero entregado: "))
# Cálculo de datos y salida de datos
monto_descuento = monto * descuento / 100
total_pagar = monto - monto_descuento
cambio = pago - total_pagar
print("Descuento: S/", round(monto_descuento, 2))
print("Total a pagar: S/", round(total_pagar, 2))
if cambio >= 0:
    print("Cambio: S/", round(cambio, 2))
else:
    print("Falta pagar: S/", round(abs(cambio), 2))

```
