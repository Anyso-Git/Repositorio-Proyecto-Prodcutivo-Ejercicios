```python
# 01. Pago de llantas
# Ingreso de datos
cantidad = int(input("Ingrese cantidad de llantas: "))
# Calculo de datos
if cantidad < 5:
    precio = 150
else:
    precio = 120
total = cantidad * precio
# Salida de datos
print("Total a pagar: S/", total)
```

```python
# 02. Determinar el menor y mayor de tres números
# Ingreso de datos
a = float(input("Número 1: "))
b = float(input("Número 2: "))
c = float(input("Número 3: "))
# Calculo de datos
mayor = max(a, b, c)
menor = min(a, b, c)
# Salida de datos
print("Mayor:", mayor)
print("Menor:", menor)
```

```python
# 03. Impuesto a tributar
# Ingreso de datos
edad = int(input("Edad: "))
ingreso = float(input("Ingreso mensual: "))
# Calculo y salida de datos
if edad > 18 and ingreso >= 3000:
    impuesto = ingreso * 0.05
    print("Debe tributar")
    print("Monto a tributar: S/", impuesto)
else:
    print("No debe tributar")
```

```python
# 04. Pago de colegiatura
# Ingreso de datos
materias = int(input("Cantidad de materias: "))
promedio = float(input("Promedio: "))
# Calculo y salida de datos
costo_total = materias * 100
if promedio >= 18:
    descuento = (costo_total / 2) * 0.50
else:
    descuento = 0
total_pagar = costo_total - descuento
print("Costo total: S/", costo_total)
print("Descuento: S/", descuento)
print("Total a pagar: S/", total_pagar)
```

```python
05. Tipo de triángulo
# Ingreso de datos
a = float(input("Lado 1: "))
b = float(input("Lado 2: "))
c = float(input("Lado 3: "))
# Calculo y salida de datos
if a == b == c:
    perimetro = a + b + c
    print("Triángulo Equilátero")
    print("Perímetro:", perimetro)

elif a == b or a == c or b == c:
    print("Triángulo Isósceles")

    if a == b:
        print("Lados iguales:", a, "y", b)
    elif a == c:
        print("Lados iguales:", a, "y", c)
    else:
        print("Lados iguales:", b, "y", c)

else:
    s = (a + b + c) / 2
    area = (s * (s-a) * (s-b) * (s-c)) ** 0.5

    print("Triángulo Escaleno")
    print("Área:", area)

```

```python
# 06. Hora N segundos después
# Ingreso de datos
hh = int(input("Horas: "))
mm = int(input("Minutos: "))
ss = int(input("Segundos: "))
n = int(input("Segundos a incrementar: "))
# Calculo y salida de datos
total_segundos = hh * 3600 + mm * 60 + ss + n
hh = (total_segundos // 3600) % 24
mm = (total_segundos % 3600) // 60
ss = total_segundos % 60
# Salida de datos
print(f"{hh:02}:{mm:02}:{ss:02}")

```

```python
# 07. Costo de llamada telefónica
# Ingreso de datos
minutos = int(input("Duración en minutos: "))
dia = input("¿Es domingo? (S/N): ").upper()
turno = input("¿Es nocturno? (S/N): ").upper()
# Calculo
if minutos <= 5:
    costo = minutos * 1.00
elif minutos <= 8:
    costo = 5 + (minutos - 5) * 0.80
elif minutos <= 10:
    costo = 5 + 2.4 + (minutos - 8) * 0.70
else:
    costo = 5 + 2.4 + 1.4 + (minutos - 10) * 0.50
if dia == "S":
    if turno == "S":
        impuesto = costo * 0.05
    else:
        impuesto = costo * 0.03
else:
    impuesto = 0
total = costo + impuesto
# Salida de Datos
print("Costo llamada: S/", round(costo, 2))
print("Impuesto: S/", round(impuesto, 2))
print("Total: S/", round(total, 2))
```

```python
# 08. Ordenar tres números
# Ingreso de datos
n1 = int(input("Número 1: "))
n2 = int(input("Número 2: "))
n3 = int(input("Número 3: "))
# Calculo  y salida de datos
if n1 == n2 or n1 == n3 or n2 == n3:
    print("Los números deben ser diferentes")
else:
    orden = input("A=Ascendente / D=Descendente: ").upper()

    numeros = [n1, n2, n3]

    if orden == "A":
        numeros.sort()
    else:
        numeros.sort(reverse=True)

    print(numeros)
```

```python
# 09. Compra de camisas
# Ingreso de datos
cantidad = int(input("Cantidad de camisas: "))
precio = float(input("Precio por camisa: "))
# Calculo de datos
subtotal = cantidad * precio
if cantidad <= 4:
    descuento = subtotal * 0.125
elif cantidad <= 8:
    descuento = subtotal * 0.20

else:
    descuento = subtotal * 0.315
total = subtotal – descuento
# Salida de datos
print("Compra sin descuento: S/", subtotal)
print("Descuento: S/", descuento)
print("Compra final: S/", total)
```

```python
# 10. Llamadas internacionales
# Ingreso de datos
clave = int(input("Ingrese código de zona: "))
minutos = int(input("Ingrese minutos: "))
# Calculo y salida de datos
if clave == 10:
    zona = "América del Norte"
    costo = 2.2
elif clave == 12:
    zona = "América del Centro"
    costo = 2.5
elif clave == 20:
    zona = "América del Sur"
    costo = 1.2
elif clave == 22:
    zona = "Asia"
    costo = 3.5
elif clave == 30:
    zona = "Europa"
    costo = 3.0
elif clave == 32:
    zona = "África"
    costo = 3.2
else:
    print("Código no válido")
    exit()
total = minutos * costo
print("Zona:", zona)
print("Costo por minuto: S/", costo)
print("Total a pagar: S/", total)

```
