```python
# 01. Suma y promedio de números pares e impares de 1 hasta N
# Definición de variables
suma_pares = 0
suma_impares = 0
cont_pares = 0
cont_impares = 0
#Ingreso de Datos
n = int(input("Ingrese N: "))
# Calculo de datos
for i in range(1, n + 1):
    if i % 2 == 0:
        suma_pares += i
        cont_pares += 1
    else:
        suma_impares += i
        cont_impares += 1
prom_pares = suma_pares / cont_pares
prom_impares = suma_impares / cont_impares
# Salida de datos
print("Suma pares:", suma_pares)
print("Promedio pares:", prom_pares)
print("Suma impares:", suma_impares)
print("Promedio impares:", prom_impares)
```

```python
# 02. Contar pares, impares, positivos, negativos y neutros
# Definición de variables
pares = impares = positivos = negativos = neutros = 0
# Calculo de datos
for i in range(10):
    num = int(input(f"Ingrese número {i+1}: "))
    if num % 2 == 0:
        pares += 1
    else:
        impares += 1
    if num > 0:
        positivos += 1
    elif num < 0:
        negativos += 1
    else:
        neutros += 1
# Salida de datos
print("Pares:", pares)
print("Impares:", impares)
print("Positivos:", positivos)
print("Negativos:", negativos)
print("Neutros:", neutros)

```

```python
# 03. Porcentajes respecto a X
# Ingreso de datos
n = int(input("Cantidad de números: "))
x = float(input("Valor de X: "))
# Calculo de datos
menores = mayores = iguales = entre = ninguna = 0
for i in range(n):
    num = float(input("Número: "))
    if num < x:
        menores += 1
    elif num > x:
        mayores += 1
    else:
        iguales += 1
    if 1 <= num <= x:
        entre += 1
    else:
        ninguna += 1
# Salida de Datos
print("Menores:", menores * 100 / n, "%")
print("Mayores:", mayores * 100 / n, "%")
print("Iguales:", iguales * 100 / n, "%")
print("Entre 1 y X:", entre * 100 / n, "%")
print("Ninguna condición:", ninguna * 100 / n, "%")
```

```python
# 04. Máximo y mínimo de N números
# Definicion de variables e Ingreso de datos
n = int(input("Cantidad de números: "))
num = float(input("Número 1: "))
mayor = menor = num
# Calculo de datos
for i in range(2, n + 1):
    num = float(input(f"Número {i}: "))
    if num > mayor:
        mayor = num
    if num < menor:
        menor = num
# Salida de datos
print("Mayor:", mayor)
print("Menor:", menor)
```

```python
# 07. Factorial de dos números
# Inicialización de variables
fact1 = 1
fact2 = 1
cadena1 = ""
cadena2 = ""
# Ingreso de datos
n1 = int(input("Primer número: "))
n2 = int(input("Segundo número: "))
# Procesamiento de datos
for i in range(1, n1 + 1):
    fact1 *= i
    cadena1 += str(i)
    if i < n1:
        cadena1 += " x "
for i in range(1, n2 + 1):
    fact2 *= i
    cadena2 += str(i)
    if i < n2:
        cadena2 += " x "
# Salida de datos
print(f"{n1}! = {cadena1} = {fact1}")
print(f"{n2}! = {cadena2} = {fact2}")
print("Son iguales hasta el múltiplo:",
      min(n1, n2))
```

```python
# 08. Control de asistentes a una fiesta
# Inicialización de variables
cantidad = 0
hombres = 0
mujeres = 0
suma_hombres = 0
suma_mujeres = 0
edad_mayor = 0
edad_menor = 999
# Procesamiento de datos
while True:
    edad = int(input("Edad: "))
    sexo = input("Sexo (M/F): ").upper()
    cantidad += 1
    if sexo == "M":
        hombres += 1
        suma_hombres += edad
    else:
        mujeres += 1
        suma_mujeres += edad
    if edad > edad_mayor:
        edad_mayor = edad
    if edad < edad_menor:
        edad_menor = edad
    resp = input("¿Registrar otra persona? (S/N): ").upper()
    if resp == "N":
        break
prom_hombres = suma_hombres / hombres if hombres > 0 else 0
prom_mujeres = suma_mujeres / mujeres if mujeres > 0 else 0
#salida de datos
print("Asistentes:", cantidad)
print("Hombres:", hombres)
print("Mujeres:", mujeres)
print("Promedio hombres:", prom_hombres)
print("Promedio mujeres:", prom_mujeres)
print("Edad mayor:", edad_mayor)
print("Edad menor:", edad_menor)
```

```python
# 09. Producto 100 × 98 × 96 × ... × 2
# Definición del vector
numeros = []
#Procesamiento de datos y salida de datos
for i in range(100, 0, -2):
    numeros.append(i)
producto = 1
for num in numeros:
    producto *= num
print("Producto =", producto)

```

```python
10. Análisis de notas de estudiantes
# Inicialización de variables
pares = 0
impares = 0
aprobados = 0
desaprobados = 0
# Ingreso de datos
n = int(input("Cantidad de estudiantes (5-20): "))
while n < 5 or n > 20:
    n = int(input("Ingrese valor entre 5 y 20: "))
notas = []
# Procesamiento de datos
for i in range(n):
    nota = float(input(f"Nota {i+1}: "))
    while nota < 0 or nota > 20:
        nota = float(input("Ingrese nota válida (0-20): "))
    notas.append(nota)
mayor = max(notas)
menor = min(notas)
cant_mayor = notas.count(mayor)
cant_menor = notas.count(menor)
promedio = sum(notas) / n
for nota in notas:
    if int(nota) % 2 == 0:
        pares += 1
    else:
        impares += 1
for nota in notas:
    if nota >= 13:
        aprobados += 1
    else:
        desaprobados += 1
porc_aprobados = aprobados * 100 / n
porc_desaprobados = desaprobados * 100 / n

#Salida de resultados
print("Mayor nota:", mayor)
print("Menor nota:", menor)
print("Cantidad con mayor nota:", cant_mayor)
print("Cantidad con menor nota:", cant_menor)
print("Promedio:", promedio)
print("Notas pares:", pares)
print("Notas impares:", impares)
print("Aprobados:", aprobados)
print("Desaprobados:", desaprobados)
print("Porcentaje aprobados:", porc_aprobados, "%")
print("Porcentaje desaprobados:", porc_desaprobados, "%")
```
